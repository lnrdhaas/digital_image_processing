## Compute the covariance and correlation matrices

```python

from pathlib import Path

# === Helper functions ===
def load_stack(files):
    """Load .npy bands, flatten, and stack into (n_pixels, n_bands)."""
    arrays = []
    shapes = set()
    for f in files:
        a = np.load(f).astype(np.float64)
        arrays.append(a)
        shapes.add(a.shape)
    if len(shapes) != 1:
        raise ValueError(f"Bands have different shapes: {shapes}")
    X = np.stack([a.ravel() for a in arrays], axis=1)
    mask = np.isfinite(X).all(axis=1)
    return X[mask]

def print_matrix(M, labels, title):
    """Pretty print matrix with labels."""
    print(f"\n=== {title} ===")
    w = max(6, max(len(s) for s in labels) + 1)
    print("".ljust(w), *[s.rjust(w) for s in labels], sep="")
    for name, row in zip(labels, M):
        print(name.ljust(w), *[f"{v:>{w}.3f}" for v in row], sep="")


root = Path("/content/")  # current directory in Colab (adjust if using Drive)

landsat_files = [
    "Data/Landsat-TM1.npy", "Data/Landsat-TM2.npy", "Data/Landsat-TM3.npy",
    "Data/Landsat-TM4.npy", "Data/Landsat-TM5.npy", "Data/Landsat-TM7.npy"
]
landsat_labels = ["TM1","TM2","TM3","TM4","TM5","TM7"]

s2_files = [
    "Data/S2-B2.npy", "Data/S2-B3.npy", "Data/S2-B4.npy", "Data/S2-B5.npy",
    "Data/S2-B6.npy", "Data/S2-B7.npy", "Data/S2-B8.npy", "Data/S2-B8A.npy",
    "Data/S2-B11.npy", "Data/S2-B12.npy"
]
s2_labels = ["B2","B3","B4","B5","B6","B7","B8","B8A","B11","B12"]


# === Landsat TM ===
X_landsat = load_stack(landsat_files)
cov_landsat = np.cov(X_landsat, rowvar=False)
corr_landsat = np.corrcoef(X_landsat, rowvar=False)

print_matrix(cov_landsat, landsat_labels, "Landsat TM Covariance Matrix")
print_matrix(corr_landsat, landsat_labels, "Landsat TM Correlation Matrix")


# Optionally save as CSVs
np.savetxt("landsat_cov.csv", cov_landsat, delimiter=",")
np.savetxt("landsat_corr.csv", corr_landsat, delimiter=",")
print("\nSaved CSV files to current directory.")


X_s2 = load_stack(s2_files)

cov_s2 = np.cov(X_s2, rowvar=False)
corr_s2 = np.corrcoef(X_s2, rowvar=False)

# Convert to DataFrames for nice formatting
cov_df = pd.DataFrame(cov_s2, index=s2_labels, columns=s2_labels)
corr_df = pd.DataFrame(corr_s2, index=s2_labels, columns=s2_labels)

# Round and display
print("\n=== Sentinel-2 Covariance Matrix ===")
print(cov_df.round(3).to_string())

print("\n=== Sentinel-2 Correlation Matrix ===")
print(corr_df.round(3).to_string())
# Your code here
```

```python
import matplotlib.pyplot as plt


s2_map = dict(zip(s2_labels, s2_files))

def load_band(label):
    return np.load(s2_map[label]).astype(np.float64)

def stretch(b, p=(2, 98)):
    lo, hi = np.percentile(b[np.isfinite(b)], p)
    b = np.clip((b - lo) / (hi - lo + 1e-12), 0, 1)
    return b


rgb_true  = np.dstack([stretch(load_band("B4")),
                       stretch(load_band("B3")),
                       stretch(load_band("B2"))])

rgb_cir   = np.dstack([stretch(load_band("B8")),   
                       stretch(load_band("B4")),  
                       stretch(load_band("B3"))])  

rgb_swir  = np.dstack([stretch(load_band("B12")),  
                       stretch(load_band("B8A")),  
                       stretch(load_band("B4"))])  

fig, axs = plt.subplots(1, 3, figsize=(16, 6))
axs[0].imshow(rgb_true); axs[0].set_title("True Color (B4-B3-B2)"); axs[0].axis("off")
axs[1].imshow(rgb_cir);  axs[1].set_title("Color IR (B8-B4-B3)");  axs[1].axis("off")
axs[2].imshow(rgb_swir); axs[2].set_title("SWIR False Color (B12-B8A-B4)"); axs[2].axis("off")
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import rasterio
import matplotlib.pyplot as plt

# Morphology
from skimage.morphology import disk, binary_erosion, binary_dilation

import ee
import geemap

plt.rcParams['figure.figsize'] = (10, 10)


project = "kth-digital-image-processing"
ee.Authenticate()
ee.Initialize(project = project)

print("EE initialized.")

```

## Load the rasters

```python

# File paths (assumed to be in the current working directory)
img_fp = "Data/paliasdes_s2.tif"              # B12, B8, B4 (in this order)
mask_fp = "Data/palisades_perimeter.tif"      # 0/1 burned area perimeter

# Read imagery
with rasterio.open(img_fp) as src:
    img = src.read()            # shape: (3, H, W)
    img_meta = src.meta.copy()
    transform = src.transform
    bounds = src.bounds

# Read mask
with rasterio.open(mask_fp) as src:
    mask = src.read(1)          # shape: (H, W)
    mask_meta = src.meta.copy()

# Ensure binary
mask = (mask > 0).astype(np.uint8)

img.shape, mask.shape
```

## Visualize the image and perimeter

```python

def percentile_stretch(a, p=(2, 98)):
    a = a.astype(np.float32)
    lo, hi = np.percentile(a[np.isfinite(a)], p)
    a = np.clip(a, lo, hi)
    denom = (hi - lo) if (hi - lo) != 0 else 1.0
    return (a - lo) / denom

# SWIR-NIR-Red false color: [B12, B8, B4] -> RGB
b12 = percentile_stretch(img[0])
b08 = percentile_stretch(img[1])
b04 = percentile_stretch(img[2])
rgb = np.stack([b12, b08, b04], axis=-1)

fig, ax = plt.subplots()
ax.imshow(rgb)
ax.imshow(np.ma.masked_where(mask == 0, mask), alpha=0.65)
ax.set_title("SWIRâ€“NIRâ€“Red composite with burned area mask overlay")
ax.axis("off")
plt.show()
```

```python

# Structuring element
selem = disk(3)

eroded = binary_erosion(mask.astype(bool), selem)
dilated = binary_dilation(mask.astype(bool), selem)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(mask, interpolation='nearest')
axs[0].set_title("Original mask")
axs[1].imshow(eroded, interpolation='nearest')
axs[1].set_title("Eroded")
axs[2].imshow(dilated, interpolation='nearest')
axs[2].set_title("Dilated")
for a in axs: a.axis("off")
plt.tight_layout()
plt.show()

print("Original pixels:", int(mask.sum()))
print("Eroded pixels:  ", int(eroded.sum()))
print("Dilated pixels: ", int(dilated.sum()))
```

Implement **opening** and **closing** with the same structuring element:
- **Opening** = erosion â†’ dilation
- **Closing** = dilation â†’ erosion

Is erosionâ†’dilation the same as dilationâ†’erosion on this mask? Try both and compare.


```python
# TODO: Implement opening and closing.

def opening(binary_mask: np.ndarray, selem) -> np.ndarray:
    """Return the opened mask (erosion then dilation)."""
    eroded = binary_erosion(binary_mask.astype(bool), selem)
    dilated = binary_dilation(eroded.astype(bool),selem)
    
    return dilated

def closing(binary_mask: np.ndarray, selem) -> np.ndarray:
    """Return the closed mask (dilation then erosion)."""
    dilated = binary_dilation(binary_mask.astype(bool), selem)
    eroded = binary_erosion(dilated.astype(bool),selem)
    
    return eroded
```

```python

# After you implement opening/closing above, run the following to compare:

opened = opening(mask.astype(bool), selem)
closed = closing(mask.astype(bool), selem)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(mask, interpolation='nearest'); axs[0].set_title("Original"); axs[0].axis("off")
axs[1].imshow(opened, interpolation='nearest'); axs[1].set_title("Opened"); axs[1].axis("off")
axs[2].imshow(closed, interpolation='nearest'); axs[2].set_title("Closed"); axs[2].axis("off")
plt.tight_layout(); plt.show()

print("Opened pixels:", int(opened.sum()))
print("Closed pixels: ", int(closed.sum()))

# Optional: look at differences
diff = closed.astype(int) - opened.astype(int)
plt.figure(figsize=(6, 6)); plt.imshow(diff, interpolation='nearest'); plt.title("Closed - Opened"); plt.axis("off"); plt.show()
```
